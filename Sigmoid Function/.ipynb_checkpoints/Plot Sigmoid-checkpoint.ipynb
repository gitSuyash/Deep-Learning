## Why regularization is necessary??

When we calculate slope for the calculation of gradient descent then if there is a rapid change in slope, there's a high probability that the loss function is left devoid of the minima. To avoid this we need to take small and convenient steps towards the minima.
<br>
For this purpose we need slope to change gradually .

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_ = range(-100000,100000)

def plot_sigmoid(slope):
    linear_output = []
    sigmoid_output = []
    for x_cord in x_:
        y_cord = slope*x_cord+5
        linear_output.append(y_cord)
        sig_output = 1/(1+np.exp(-y_cord))
        sigmoid_output.append(sig_output)
        
    plt.plot(linear_output,sigmoid_output)
    plt.grid(True)
```

```python
plot_sigmoid(0.001)
```

```python
plot_sigmoid(0.0005)
```

```python
plot_sigmoid(0.0002)
```

```python
plot_sigmoid(0.0001)
```
